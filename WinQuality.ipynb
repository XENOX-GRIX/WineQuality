```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
wine=pd.read_csv("winequalityN.csv")
wine.describe
```

```python
wine.dropna(inplace=True)
wine.isna().sum()
```

```python
wine.describe
```

### Type column has non-numeric type of data , so converting it to numeric 

```python
dum = pd.get_dummies(wine["type"])
y=wine["quality"]
x=wine.drop("quality",axis=1)
x=x.join(dum)
x=x.drop("type",axis=1)
```

```python
wine.head()
x.head()
```

```python
sns.regplot(x="fixed acidity", y="quality", data=wine);
```

```python
sns.pairplot(wine,vars=["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol"]);
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(wine.corr(),annot=True);
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm 
```

```python
modelsvr=svm.SVR()
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size =0.2)
```

```python
modelsvr.fit(x_train,y_train)
```

```python
modelsvr.score(x_test,y_test)
```
